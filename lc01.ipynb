```python
# Don't just name the loader, tell it to LOAD
loader = PyPDFLoader("lctest_for_pdf.pdf")
docs = loader.load() # This creates a LIST of pages
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=100, chunk_overlap=0
)

# Now 'docs' is a list we can iterate through
lc_full_text = "\n".join([doc.page_content for doc in docs])
lc_texts = text_splitter.split_text(lc_full_text)

print(lc_texts)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the "Smart" Splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20, # This gives the AI some 'context' from the previous chunk
    length_function=len,
    is_separator_regex=False,
)

# Turn your docs into chunks
# We use .split_documents instead of .split_text to keep the metadata (like filename)
chunks = text_splitter.split_documents(docs)

print(f"Number of chunks created: {len(chunks)}")
print(f"Content of first chunk:\n{chunks[0].page_content}")
```
